```python
from selenium import webdriver
import pandas as pd
import time

def scrape_jobs():
    # Step 1: Set up the Selenium webdriver
    driver = webdriver.Chrome("path-to-chrome-driver/chromedriver") # Replace with the path to your ChromeDriver executable
    driver.get("https://www.shine.com/")

    # Step 2: Enter search criteria and click search
    job_title = "Data Analyst"
    location = "Bangalore"

    job_title_input = driver.find_element_by_id("search-keyword")
    location_input = driver.find_element_by_id("loc")
    submit_button = driver.find_element_by_class_name("search-button")

    job_title_input.send_keys(job_title)
    location_input.send_keys(location)
    submit_button.click()

    # Wait for the page to load
    time.sleep(5)

    # Step 3: Scrape the data for the first 10 jobs
    jobs = []
    job_results = driver.find_elements_by_class_name("search_listing")
    for result in job_results[:10]:
        job_title = result.find_element_by_class_name("job_title").text.strip()
        job_location = result.find_element_by_class_name("location").text.strip()
        company_name = result.find_element_by_class_name("company").text.strip()
        experience_required = result.find_element_by_class_name("exp").text.strip()

        jobs.append({
            'Job Title': job_title,
            'Job Location': job_location,
            'Company Name': company_name,
            'Experience Required': experience_required
        })

    # Step 4: Create a dataframe of the scraped data
    df = pd.DataFrame(jobs)

    # Step 5: Quit the webdriver
    driver.quit()

    return df

# Running the function to scrape the jobs
job_dataframe = scrape_jobs()
print(job_dataframe)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Initialize the Chrome driver
driver_path = 'path_to_your_webdriver_executable'
driver = webdriver.Chrome(executable_path=driver_path)

# Open the Flipkart website
driver.get("https://www.flipkart.com/")

# Close the login popup if it appears
try:
    close_button = driver.find_element(By.CSS_SELECTOR, "button._2KpZ6l._2doB4z")
    close_button.click()
except:
    pass

# Find the search input field and enter "sunglasses"
search_input = driver.find_element(By.CSS_SELECTOR, "input.LM6RPg")
search_input.send_keys("sunglasses")
search_input.submit()

# Initialize variables for scraped data
sunglasses_data = []

# Loop until we have data for 100 sunglasses
while len(sunglasses_data) < 100:
    # Wait for the page to load
    time.sleep(2)

    # Find all product containers on the page
    product_containers = driver.find_elements(By.CSS_SELECTOR, "div.bhgxx2")

    # Loop through each product container and scrape data
    for container in product_containers:
        try:
            brand = container.find_element(By.CSS_SELECTOR, "div._2WkVRV").text
            description = container.find_element(By.CSS_SELECTOR, "aIRwtW").text
            price = container.find_element(By.CSS_SELECTOR, "._1uv9Cb").text

            sunglasses_data.append({
                "Brand": brand,
                "Product Description": description,
                "Price": price
            })
        except:
            pass

        # Break the loop if we have data for 100 sunglasses
        if len(sunglasses_data) >= 100:
            break

    # Go to the next page if available
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, "a._1LKTO3")
        next_button.click()
    except:
        break

# Create a DataFrame from the scraped data
df = pd.DataFrame(sunglasses_data)

# Print the DataFrame
print(df)

# Close the browser
driver.quit()
```
