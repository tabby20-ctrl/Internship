```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"

try:
    # Send a GET request to the URL
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception if there was an error with the request

    # Create a BeautifulSoup object
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the table containing the video details
    table = soup.find('table', attrs={'class': 'wikitable'})

    # Iterate over each row in the table (skipping the header row)
    for row in table.find_all('tr')[1:]:
        # Extract the required details from each column
        columns = row.find_all('td')
        rank = columns[0].text.strip()
        name = columns[1].text.strip()
        artist = columns[2].text.strip()
        upload_date = columns[3].text.strip()
        views = columns[4].text.strip()

        # Print the details
        print("Rank:", rank)
        print("Name:", name)
        print("Artist:", artist)
        print("Upload Date:", upload_date)
        print("Views:", views)
        print("----------")

except requests.exceptions.RequestException as e:
    # Handle any exceptions that occurred during the request
    print("Error:", e)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.bcci.tv/"

try:
    # Send a GET request to the URL
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception if there was an error with the request

    # Create a BeautifulSoup object
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the link to the international fixtures page
    fixtures_link = soup.find("a", text="International Fixtures")
    fixtures_url = url + fixtures_link["href"]

    # Send a GET request to the fixtures URL
    fixtures_response = requests.get(fixtures_url)
    fixtures_response.raise_for_status()  # Raise an exception if there was an error with the request

    # Create a BeautifulSoup object for the fixtures page
    fixtures_soup = BeautifulSoup(fixtures_response.text, 'html.parser')

    # Find the table containing the fixture details
    fixtures_table = fixtures_soup.find('div', attrs={'class': 'js-list'})

    # Iterate over each fixture card
    for fixture_card in fixtures_table.find_all('div', attrs={'class': 'fixture-widget'}):
        # Extract the required details from the card
        match_title = fixture_card.find('span', attrs={'class': 'fixture__format'}).text.strip()
        series = fixture_card.find('span', attrs={'class': 'u-unskewed-text'}).text.strip()
        place = fixture_card.find('p', attrs={'class': 'fixture__additional-info'}).text.strip()
        date = fixture_card.find('span', attrs={'class': 'fixture__date'}).text.strip()
        time = fixture_card.find('span', attrs={'class': 'fixture__time'}).text.strip()

        # Print the details
        print("Match Title:", match_title)
        print("Series:", series)
        print("Place:", place)
        print("Date:", date)
        print("Time:", time)
        print("----------")

except requests.exceptions.RequestException as e:
    # Handle any exceptions that occurred during the request
    print("Error:", e)
```

```python
import requests
from bs4 import BeautifulSoup

url = "http://statisticstimes.com/"

try:
    # Send a GET request to the URL
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception if there was an error with the request

    # Create a BeautifulSoup object
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the link to the economy page
    economy_link = soup.find("a", text="Economy")
    economy_url = url + economy_link["href"]

    # Send a GET request to the economy URL
    economy_response = requests.get(economy_url)
    economy_response.raise_for_status()  # Raise an exception if there was an error with the request

    # Create a BeautifulSoup object for the economy page
    economy_soup = BeautifulSoup(economy_response.text, 'html.parser')

    # Find the link to the State-wise GDP page
    gdp_link = economy_soup.find("a", text="GDP of Indian states")
    gdp_url = url + gdp_link["href"]

    # Send a GET request to the GDP URL
    gdp_response = requests.get(gdp_url)
    gdp_response.raise_for_status()  # Raise an exception if there was an error with the request

    # Create a BeautifulSoup object for the GDP page
    gdp_soup = BeautifulSoup(gdp_response.text, 'html.parser')

    # Find the table containing the GDP details
    gdp_table = gdp_soup.find('table', attrs={'id': 'table_id'})

    # Iterate over each row in the table (skipping the header row)
    for row in gdp_table.find_all('tr')[1:]:
        # Extract the required details from each column
        columns = row.find_all('td')
        rank = columns[0].text.strip()
        state = columns[1].text.strip()
        gdp_18_19 = columns[2].text.strip()
        gdp_19_20 = columns[3].text.strip()
        share_18_19 = columns[4].text.strip()
        gdp_billion = columns[5].text.strip()

        # Print the details
        print("Rank:", rank)
        print("State:", state)
        print("GSDP(18-19) - at current prices:", gdp_18_19)
        print("GSDP(19-20) - at current prices:", gdp_19_20)
        print("Share(18-19):", share_18_19)
        print("GDP($ billion):", gdp_billion)
        print("----------")

except requests.exceptions.RequestException as e:
    # Handle any exceptions that occurred during the request
    print("Error:", e)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://github.com/"

try:
    # Send a GET request to the URL
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception if there was an error with the request

    # Create a BeautifulSoup object
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the Explore menu
    explore_menu = soup.find("nav", class_="mt-0 px-3 px-lg-0 mb-5 mb-lg-0")

    # Find the link to the trending page
    trending_link = explore_menu.find("a", href="/trending")
    trending_url = url + trending_link["href"]

    # Send a GET request to the trending URL
    trending_response = requests.get(trending_url)
    trending_response.raise_for_status()  # Raise an exception if there was an error with the request

    # Create a BeautifulSoup object for the trending page
    trending_soup = BeautifulSoup(trending_response.text, 'html.parser')

    # Find the list of trending repositories
    repositories = trending_soup.find_all("article", class_="Box-row")

    # Iterate over each repository
    for repository in repositories:
        # Extract the required details
        title = repository.find("h1", class_="h3 lh-condensed").text.strip()
        description = repository.find("p", class_="col-9 color-text-secondary my-1 pr-4").text.strip()
        contributors = repository.find("a", class_="Link--muted d-inline-block mr-3").text.strip()
        language = repository.find("span", itemprop="programmingLanguage").text.strip()

        # Print the details
        print("Repository title:", title)
        print("Repository description:", description)
        print("Contributors count:", contributors)
        print("Language used:", language)
        print("----------")

except requests.exceptions.RequestException as e:
    # Handle any exceptions that occurred during the request
    print("Error:", e)
```
